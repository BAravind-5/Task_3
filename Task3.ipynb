```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/products_missing_demo (1).csv")
df.head()
```

```python

numerical_cols = ['Price', 'Units_Sold', 'Revenue', 'Trend_Score']

df[numerical_cols].hist(figsize=(10,6), bins=10)
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

Price and Revenue show right-skewed distributions.
Trend_Score is fairly normally distributed.

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Category')
plt.xticks(rotation=45)
plt.title("Product Count by Category")
plt.show()
```

Electronics and Fashion categories dominate the dataset.

```python

plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Price')
plt.title("Box Plot of Price")
plt.show()
```

A few high-priced products appear as outliers, which may influence mean values.

```python

plt.figure(figsize=(6,5))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Revenue is strongly correlated with Units_Sold and Price, making them important
features for prediction.

Important Features for Prediction:
- Units_Sold (strong impact on Revenue)
- Price
- Trend_Score

EDA Summary:
- Numerical features show varied distributions.
- Some outliers exist in Price and Revenue.
- Electronics is the most common category.
- Revenue strongly depends on Units_Sold and Price.
- Visualization helped identify patterns and relationships clearly.
